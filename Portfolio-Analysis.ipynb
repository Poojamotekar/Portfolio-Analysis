### 3.2 Plot the historical cumulative returns of the portfolio.

### Formula:cumulative_returns=(Current Value - Initial Value) / Initial Value

```python
# Calculate daily returns (in weights %) of the portfolio
weights_returns = (daily_returns_assets * portfolio_weights).sum(axis=1)
```

```python
# Cumulative returns
cumulative_returns = (1 + weights_returns).cumprod()
```

```python
cumulative_returns
```

```python
portfolio_daily_returns = (daily_returns_assets * portfolio_weights).sum(axis=1)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(cumulative_returns.index, cumulative_returns, color='blue', linestyle='-')
plt.title('Historical Cumulative Returns of the Portfolio')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.grid(True)
plt.show()
```

#The direction of the line indicates whether the portfolio's value has been increasing or decreasing over the specified time period. An upward trend indicates positive returns, while a downward trend suggests negative returns.
how quickly the line on a graph rises or falls. A steep line indicates a rapid change, while a less steep line suggests a slower change.

### 3.3Calculate the annualized return of the portfolio.

```python
# Step 1: Calculate the total return
portfolio_value_start = cumulative_returns.iloc[0]  # Initial portfolio value
portfolio_value_end = cumulative_returns.iloc[-1]  # Final portfolio value
total_return = (portfolio_value_end - portfolio_value_start) / portfolio_value_start

# Step 2: Determine the investment period
start_date = cumulative_returns.index[0]  # Start date of the investment
end_date = cumulative_returns.index[-1]   # End date of the investment
investment_period_days = (end_date - start_date).days
investment_period_years = investment_period_days / 365

# Step 3: Adjust to annual time
annualized_return = (((1 + total_return) ** (1 / investment_period_years)) - 1).round(2)

# Print the annualized return
print(f"Portfolio's Annual return: {annualized_return}")
```
